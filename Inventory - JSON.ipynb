```python
#Importing required libraries.
import json
import time 

# Loading invetory data
sale = ''

fd=open('Records.json','r')
js=fd.read()
fd.close() 

record= json.loads(js)

record= {1001:{"Name":"Dairy Milk",'Price':10,'Qn':400},
         1002:{"Name":"Candy",'Price':5,'Qn':1000},
         1003:{"Name":"Chocolate Bar",'Price':25,'Qn':500},
         1004:{"Name":"Cake",'Price':50,'Qn':600},
         1005:{"Name":"Icepops",'Price':30,'Qn':700}}

# Displaying the menu
print("-----------MENU---------------\n")


for key in record.keys():
    print(key,":", record[key]['Name'],"|",record[key]['Price'],"|",record[key]['Qn'])
    
print("------------------------------\n")

# Taking the user input
u_name=str(input("Enter your name         :"))
u_email=str(input("Enter your Email       :"))
u_mob_no=str(input("Enter your Mob. No.   :"))
u_prod=int(input("Enter Product ID        : " ))
u_qn  =int(input("Enter Product Quantity  : "))

# Processing the purchase
if (record[u_prod]['Qn'] >= u_qn):
    

    record[u_prod]['Qn']=record[u_prod]['Qn']- u_qn
    print("---------------------------------------")

    print("-----------------Bill-------------------\n")

    print("Name           :"  ,record[u_prod]['Name'])
    print("Price          :"  ,record[u_prod]['Price'], "Rs")
    print("Quantity       :"  ,u_qn)

    print("------------------------------")
    print("Billing Amount :" ,record[u_prod]['Price'] * u_qn, "Rs" )
    print("------------------------------")
    
    sale ='1'+","+u_name+","+u_mob_no+","+u_email+","+record[u_prod]['Name']+","+str(u_qn)+","+str(record[u_prod]['Price'])+","+str(record[u_prod]['Qn'] * record[u_prod]['Price'])+","+ time.ctime()+"\n"

    
else:
    print("Sorry, We're not having enough quantity in our inventory.")
    print("We're only having", record[u_prod]['Qn'], 'quantity.')
    print("------------------------------------------------------")
    
    ch=str(input("Press Y/y to Purchase:  "))
    
    if (ch == "Y" or ch == "y"):
        
        print("-----------Bill-----------")
        print("Name           :"  ,record[u_prod]['Name'])
        print("Price          :"  ,record[u_prod]['Price'], "Rs")
        print("Quantity       :"  ,record[u_prod]['Qn'])

        print("------------------------------")
        print("Billing Amount :" ,record[u_prod]['Price'] * record[u_prod]['Qn'], "Rs")
        print("------------------------------")
    
        sale ='1'+","+u_name+","+u_mob_no+","+u_email+","+record[u_prod]['Name']+","+str(record[u_prod]['Qn'])+","+str(record[u_prod]['Price'])+","+str(record[u_prod]['Qn'] * record[u_prod]['Price'])+","+ time.ctime()+"\n"

        record[u_prod]["Qn"]=0

    else:
        print("Thanks!")
        
     
#Updating the inventory.
js=json.dumps(record)

fd=open('Records.json','w')
fd.write(js)
fd.close() 

# Recording the sale.
fd=open('Sales.txt','a')
fd.write(sale)
fd.close()


print('')
print("--------------------------------------------")
print("Thankyou for your order! Inventory Updated.")
print("--------------------------------------------")
```
